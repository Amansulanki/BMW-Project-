```python
# Sales_Classification by categories 
a = df['Sales_Classification'].value_counts()
categories = ['Low', 'High']
value = [34754,15246]
colors = ['blue', 'orange']
plt.bar(categories, value, color=colors)
plt.title('Sales Classification', fontsize=14, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.show()
```

```python

colors = ['blue', 'orange']

sales_class = df.groupby(['Model', 'Sales_Classification'])['Sales_Volume'].mean().unstack(fill_value=0)


sales_class.plot(
    kind='bar',
    figsize=(12,6),
    color=colors
)

plt.title('Average Sales Volume by Model and Sales Classification', fontsize=14, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Average Sales Volume', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Total Sales Volume by Model
import matplotlib.pyplot as plt
colors_dark = [
    '#14406d',  
    '#cc5f00',  
    '#1f6b1f',  
    '#8b0000',  
    '#5a3a99',  
    '#5a3b2f',  
    '#a74f88',  
    '#4c4c4c',  
    '#7d7b16', 
    '#137a86',  
    '#b3403a'   
]


df.groupby('Model')['Sales_Volume'].mean().sort_values().plot(
    kind='bar',
    figsize=(10,6),
    color=colors
)
plt.title('Total Sales Volume by Model', fontsize=14, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Sale Volume', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Car Models by Fuel Type

fuel_model = df.groupby(['Model', 'Fuel_Type']).size().unstack(fill_value=0)
fuel_model.plot(kind='bar', figsize=(10,6), width=0.8)
plt.title('Car Models by Fuel Type', fontsize=14, fontweight='bold')
plt.xlabel('Car Model', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Fuel Type')
plt.tight_layout()
plt.show()
```

```python
# Sales_Classification by Transmission

b = df.groupby(['Transmission', 'Sales_Classification']).size().unstack(fill_value=0)
b.plot(kind='bar', figsize=(10,6), width=0.8)
plt.title('Transmission vs Sales Classification', fontsize=14, fontweight='bold')
plt.xlabel('Transmission Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Sales Classification')
plt.tight_layout()
plt.show()
```

```python
# Car colors by model

c = df.groupby(['Model', 'Color']).size().unstack(fill_value=0)
c.plot(kind='bar', figsize=(10,6), width=0.7)
plt.title('Car Models by Color', fontsize=14, fontweight='bold')
plt.xlabel('Model', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='color')
plt.tight_layout()
plt.show()
```

```python
# Total Sales Volume by year
df.groupby('Year')['Sales_Volume'].sum().plot(
    kind='line',
    figsize=(8,5),
    marker='o',      
    linewidth=2,      
    color='blue'
)


plt.title('BMW Sales Volume Over the Years', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales Volume', fontsize=12)


plt.grid(True, linestyle='--', alpha=0.6)


plt.tight_layout()
plt.show()

```

```python
# Region vs Sales Volume
import matplotlib.pyplot as plt
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']
df.groupby('Region')['Sales_Volume'].mean().sort_values().plot(
    kind='bar',
    figsize=(10,6),
    color=colors
)
plt.title('Sales volum by Region', fontsize=14, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sale Volume', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Fuel_Type vs Sales_Volume
colors = ['red', 'yellow', 'blue', 'purple']
df.groupby('Fuel_Type')['Sales_Volume'].mean().sort_values().plot(
    kind='bar',
    figsize=(10,6),
    color=colors
)
plt.title('Sales volum by Fuel Type', fontsize=14, fontweight='bold')
plt.xlabel('Fuel Type', fontsize=12)
plt.ylabel('Sale Volume', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Model vs Price_USD

colors = ['blue', 'orange', 'green', 'red', 'purple', 'gold', 'teal', 'pink']
df.groupby('Model')['Price_USD'].mean().sort_values().plot(
    kind='bar',
    figsize=(10,6),
    color=colors
)
plt.title('Average Price by BMW Model', fontsize=14, fontweight='bold')
plt.xlabel('Model Name', fontsize=12)
plt.ylabel('Average Price (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

avg_price = df.groupby('Engine_Size_L')['Price_USD'].mean().sort_index()

plt.figure(figsize=(10,6))
plt.plot(avg_price.index, avg_price.values, marker='o', color='red', linewidth=2)

plt.title('Average Price vs Engine Size', fontsize=14, fontweight='bold')
plt.xlabel('Engine Size (L)', fontsize=12)
plt.ylabel('Average Price (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


avg_price = df.groupby('Engine_Size_L')['Price_USD'].mean().sort_index()

plt.figure(figsize=(13,7))
avg_price.plot(kind='bar', color='blue', edgecolor='black')

plt.title('Average Price by Engine Size', fontsize=14, fontweight='bold')
plt.xlabel('Engine Size (L)', fontsize=12)
plt.ylabel('Average Price (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
